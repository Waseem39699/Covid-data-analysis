## ADD YEAR COLUMN, MONTH COLUMN,AND DAY COLUMN (jan-dec)

```python
df = pd.read_csv('country_wise_latest.csv')
df['date'] = pd.to_datetime(df['date'])

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['month'] = df['month'].astype(int)

df['month'] = df['month'].apply(lambda x : calendar.month_abbr[x])
df.fillna(0)
df.to_csv('country_wise_latest.csv')
df
```

## Month with the most number of total vaccinations in India in 2021

```python
india_vaccinations_2021 = df[(df['country'] == 'India') & (df['year'] == 2021)]

# Grouping data by month and calculating total vaccinations for each month in 2021
total_vaccinations_by_month_2021 = india_vaccinations_2021.groupby('month')['total_vaccinations'].sum().reset_index()

# Finding the month with the most vaccinations in 2021
most_vaccinated_month_2021 = total_vaccinations_by_month_2021.loc[total_vaccinations_by_month_2021['total_vaccinations'].idxmax()]

print("Month with the most number of total vaccinations in India in 2021:", most_vaccinated_month_2021['month'])
```
